```python
# Title: Advanced Data Visualization
# Description: Create a pair plot for the dataset to display pairwise relationships.
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Create a pair plot to visualize pairwise relationships
sns.pairplot(df, hue='species', markers=["o", "s", "D"], palette="Set2")

# Add a title and show the plot
plt.suptitle('Pair Plot of Iris Dataset Features', y=1.02)
plt.show()
```
