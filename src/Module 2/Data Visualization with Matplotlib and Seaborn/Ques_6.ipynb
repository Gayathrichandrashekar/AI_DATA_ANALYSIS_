```python
# Customizing Matplotlib Plot Styles
# Description: Customize the Matplotlib bar plot from Question 1 with a grid, different color, and custom fonts.
import matplotlib.pyplot as plt

# Data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Plotting the bar chart with customizations
plt.bar(products, sales, color='darkorange', edgecolor='black')

# Customizing the plot with a grid and custom fonts
plt.title('Product Sales Comparison', fontsize=18, fontfamily='serif', fontweight='bold')
plt.xlabel('Products', fontsize=14, fontfamily='sans-serif')
plt.ylabel('Sales (in thousands)', fontsize=14, fontfamily='sans-serif')

# Enabling grid and customizing grid lines
plt.grid(True, linestyle='--', linewidth=0.7, color='gray')

# Show the plot
plt.show()
```
