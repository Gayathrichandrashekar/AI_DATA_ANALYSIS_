```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]
sizes = [10, 15, 13, 18, 12]

# Create a DataFrame
data = pd.DataFrame({
    'Height': heights,
    'Weight': weights,
    'Size': sizes
})

# Plotting
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Height', y='Weight', size='Size', data=data, hue='Size', sizes=(50, 250), palette='viridis', legend=False)

# Titles and labels
plt.title('Height vs Weight with Point Sizes', fontsize=16)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Weight (kg)', fontsize=12)
plt.grid(True)
plt.show()
```
