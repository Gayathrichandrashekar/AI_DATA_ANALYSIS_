```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Drop the 'species' column since it is categorical and not suitable for correlation calculation
iris_numeric = iris.drop(columns=['species'])

# Compute the correlation matrix
corr_matrix = iris_numeric.corr()

# Create the heatmap with advanced styling
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, 
            vmin=-1, vmax=1, cbar_kws={'shrink': 0.75}, square=True)

# Customize the plot
plt.title("Correlation Matrix Heatmap of Iris Dataset", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
