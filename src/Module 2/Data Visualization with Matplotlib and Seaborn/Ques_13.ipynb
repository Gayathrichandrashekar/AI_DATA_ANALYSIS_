```python
# Customizing Plots with Seaborn - Boxplot
# Objective: Customize a boxplot with Seaborn to show the distribution of petal lengths by
# species.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a boxplot for petal length by species
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris, x='species', y='petal_length', palette='pastel')

# Customize the plot
plt.title("Distribution of Petal Length by Species", fontsize=16)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Petal Length (cm)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
