```python
# Advanced Seaborn Plot - Pairplot
# Objective: Use Seaborn to create a pairplot to visualize pairwise relationships in a dataset.
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Create pairplot with hue based on 'sex'
sns.pairplot(tips, hue='sex', palette='Set2')

# Display the plot
plt.suptitle("Pairplot of Tips Dataset by Gender", y=1.02, fontsize=16)
plt.show()
```
