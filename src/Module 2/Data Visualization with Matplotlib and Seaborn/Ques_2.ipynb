```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Convert to NumPy arrays for validation
heights_array = np.array(heights)
weights_array = np.array(weights)

# Validation checks
if len(heights_array) == 0 or len(weights_array) == 0:
    raise ValueError("Height and weight lists must not be empty.")
if not np.issubdtype(heights_array.dtype, np.number) or not np.issubdtype(weights_array.dtype, np.number):
    raise ValueError("Height and weight data must be numeric.")
if len(heights_array) != len(weights_array):
    raise ValueError(f"Length mismatch: {len(heights)} heights vs {len(weights)} weights.")

# Plot using Seaborn
sns.set(style='whitegrid')
sns.scatterplot(x=heights_array, y=weights_array, color='blue')
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```
