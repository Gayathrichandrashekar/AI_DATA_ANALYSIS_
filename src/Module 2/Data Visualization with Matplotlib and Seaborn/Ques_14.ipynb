```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 16]

# Create line plot with customized styles
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='teal', linestyle='--', linewidth=2, marker='o', markersize=8)

# Add labels and title
plt.title("Styled Line Plot Example", fontsize=16, fontweight='bold')
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.grid(True, linestyle=':', linewidth=0.7)
plt.tight_layout()
plt.show()
```
