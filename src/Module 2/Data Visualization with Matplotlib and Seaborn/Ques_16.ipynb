```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.
import matplotlib.pyplot as plt
import seaborn as sns

# Example data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Create a figure with 1 row and 2 columns of subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Bar plot on the first subplot
axes[0].bar(products, sales, color='skyblue')
axes[0].set_title("Product Sales")
axes[0].set_xlabel("Product")
axes[0].set_ylabel("Sales (in thousands)")

# Scatter plot on the second subplot
axes[1].scatter(heights, weights, color='orange')
axes[1].set_title("Height vs. Weight")
axes[1].set_xlabel("Height (cm)")
axes[1].set_ylabel("Weight (kg)")

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the combined plots
plt.show()
```
