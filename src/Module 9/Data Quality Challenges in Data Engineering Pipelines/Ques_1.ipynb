### Task 1: Detecting Missing Values during Data Ingestion
**Description**: You have a CSV file with missing values in some columns. Write a Python script to detect and report missing values during the ingestion process.

**Steps**:
1. Load data
2. Check for missing values
3. Report missing values

```python
# Write your code from here
# Ques_1.ipynb
# Task: Detecting Missing Values during Data Ingestion

# Step 1: Import Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Simulate or Load Data (replace with actual ingestion source if needed)
# Example: Creating sample data with missing values for demonstration
data = {
    'CustomerID': [101, 102, 103, 104, 105, None],
    'TransactionAmount': [250.0, None, 340.5, 120.0, None, 410.0],
    'TransactionDate': ['2024-01-01', '2024-01-02', None, '2024-01-04', '2024-01-05', '2024-01-06'],
    'Location': ['NY', 'CA', 'TX', None, 'TX', 'CA']
}

df = pd.DataFrame(data)

# Step 3: Display Raw Data
print("Raw Ingested Data:")
display(df)

# Step 4: Detect Missing Values
missing_summary = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

# Combine into a missing value report
missing_report = pd.DataFrame({
    'Missing Count': missing_summary,
    'Missing Percent': missing_percent
})

# Filter only columns with missing values
missing_report = missing_report[missing_report['Missing Count'] > 0]

print("\nMissing Value Report:")
display(missing_report)

# Step 5: Visualize Missing Data
plt.figure(figsize=(8, 4))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Value Heatmap")
plt.show()

# Step 6: Handle Missing Values
# Strategy:
# - Drop columns with more than 50% missing
# - Impute numeric columns with mean
# - Impute categorical columns with mode

# Drop columns with more than 50% missing values
df_cleaned = df.dropna(thresh=len(df) * 0.5, axis=1)

# Impute numeric columns with mean
numeric_cols = df_cleaned.select_dtypes(include='number').columns
for col in numeric_cols:
    df_cleaned[col].fillna(df_cleaned[col].mean(), inplace=True)

# Impute categorical columns with mode
categorical_cols = df_cleaned.select_dtypes(include='object').columns
for col in categorical_cols:
    df_cleaned[col].fillna(df_cleaned[col].mode()[0], inplace=True)

print("\nCleaned Data After Handling Missing Values:")
display(df_cleaned)

# Step 7: (Optional) Save cleaned data
# df_cleaned.to_csv('cleaned_data.csv', index=False)


```
