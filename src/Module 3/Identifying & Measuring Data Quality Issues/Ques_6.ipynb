```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

# Create synthetic dataset
np.random.seed(42)
data = {
    'age': np.random.normal(loc=35, scale=10, size=1000)  # Normal distribution around age 35
}
df = pd.DataFrame(data)

# Choose the column to analyze
column_name = 'age'

# Descriptive statistics
print("Descriptive Statistics:")
print(df[column_name].describe())
print(f"Skewness: {skew(df[column_name]):.2f}")
print(f"Kurtosis: {kurtosis(df[column_name]):.2f}")

# Plot histogram with KDE
plt.figure(figsize=(10, 6))
sns.histplot(df[column_name], bins=30, kde=True, color='skyblue', edgecolor='black')
plt.title(f'Distribution of {column_name}')
plt.xlabel(column_name)
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
