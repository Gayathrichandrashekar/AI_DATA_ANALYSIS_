```python
# Part 1: Handle Missing Values & Duplicates

    # Step-by-Step Guidelines:
# 1. Load the Data: First, ensure you have pandas installed and import it.
# 2. Handling Missing Values
#     1. Identify Missing Values:
#     2. Fill Missing Values:
# 3. Handling Duplicates
#     1. Identify Duplicates:
#     2. Remove Duplicates:
# 4. Combined Practice on a New Dataset
#     1. New Sample Data:
#     2. Handling Missing Values:
#     3. Remove Duplicates:
        
        
        
        
        # Step 1: Import the necessary library
import pandas as pd
import numpy as np

# Step 2: Load the Data
# For this example, we will create a sample dataset

data = {
    'ID': [1, 2, 3, 4, 5, 6, 7],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Alice', None],
    'Age': [25, 30, None, 35, 40, 25, 30],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'F', None],
    'Salary': [50000, None, 60000, 70000, None, 50000, 55000]
}

df = pd.DataFrame(data)

# Step 3: Handling Missing Values
# 3.1 Identify Missing Values
print("Missing Values in the DataFrame:")
print(df.isnull().sum())

# 3.2 Fill Missing Values
# Filling missing values in 'Age' with the mean of the column
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Filling missing values in 'Gender' with the mode (most frequent value)
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)

# Filling missing values in 'Salary' with the median of the column
df['Salary'].fillna(df['Salary'].median(), inplace=True)

print("\nData after handling missing values:")
print(df)

# Step 4: Handling Duplicates
# 4.1 Identify Duplicates
print("\nDuplicate Entries in the DataFrame:")
print(df.duplicated().sum())

# 4.2 Remove Duplicates
# Remove duplicate rows based on all columns
df.drop_duplicates(inplace=True)

print("\nData after removing duplicates:")
print(df)

# Step 5: Combined Practice on a New Dataset
# 5.1 New Sample Data
data_new = {
    'ID': [8, 9, 10, 11, 12, 13],
    'Name': ['Grace', 'Helen', 'Alice', 'Ivan', 'John', 'Alice'],
    'Age': [29, None, 30, 25, 33, None],
    'Gender': ['F', 'F', 'F', 'M', 'M', 'M'],
    'Salary': [65000, 70000, None, 55000, 75000, 60000]
}

df_new = pd.DataFrame(data_new)

# 5.2 Handling Missing Values
df_new['Age'].fillna(df_new['Age'].mean(), inplace=True)
df_new['Salary'].fillna(df_new['Salary'].median(), inplace=True)

# 5.3 Remove Duplicates
df_new.drop_duplicates(inplace=True)

print("\nNew Data after handling missing values and duplicates:")
print(df_new)

        
        
        
        
```
